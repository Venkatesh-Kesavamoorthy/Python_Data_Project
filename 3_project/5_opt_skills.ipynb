```python

# Importing Libraries
import ast
import seaborn as sns
import pandas as pd 
from datasets import load_dataset
import matplotlib.pyplot as plt
from adjustText import adjust_text

# Loading Data
ds = load_dataset('lukebarousse/data_jobs')
df = ds['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df
```

```python
df_da_sg = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'Singapore')].copy()

df_da_sg
```

```python
df_da_sg = df_da_sg.dropna(subset= ['salary_year_avg'])

df_da_sg_exploded = df_da_sg.explode('job_skills')

df_da_sg_exploded[['salary_year_avg','job_skills']].head(5)
```

```python
df_da_skills = df_da_sg_exploded.groupby('job_skills')['salary_year_avg'].agg(['count','median']).sort_values(by= 'count', ascending= False)

df_da_skills = df_da_skills.rename(columns={'count': "skill_count", 'median': 'median_salary'})

da_job_counts = len(df_da_sg)

df_da_skills['skill_percent'] = df_da_skills['skill_count'] / da_job_counts * 100

skill_perc = 8

df_da_skills_high_demand = df_da_skills[df_da_skills['skill_percent'] > skill_perc]

df_da_skills_high_demand
```

```python
from adjustText import adjust_text

# fig, ax = plt.subplots()


df_da_skills_high_demand.plot(kind= 'scatter', x= 'skill_percent', y='median_salary')

texts = []
for i, txt in enumerate(df_da_skills_high_demand.index):
    texts.append(plt.text(df_da_skills_high_demand['skill_percent'].iloc[i],df_da_skills_high_demand['median_salary'].iloc[i], txt))
    

plt.xlabel('Percent of Data Analysts Jobs')
plt.ylabel('Median Yearly Salary')
plt.title('Most Optimal Skills for Data Analysts in the Sg')
plt.tight_layout()
plt.xlim(0,60)

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, post: f'${int(y/1000)}k'))
ax.xaxis.set_major_formatter(PercentFormatter(decimals=0))
```

```python
df_technology = df['job_type_skills'].copy()

df_technology = df_technology.drop_duplicates()

df_technology = df_technology.dropna()

tech_dict = {}
for row in df_technology:
    row_dict = ast.literal_eval(row)
    for key, value in row_dict.items():
        if key in tech_dict:
            tech_dict[key] += value
        else:
            tech_dict[key] = value

for key,value in tech_dict.items():
    tech_dict[key] = list(set(value))

tech_dict
```

```python
pd.DataFrame(list(tech_dict.items()), columns=['technology', 'skills'])

df_technology = pd.DataFrame(list(tech_dict.items()), columns=['technology', 'skills'])

df_technology = df_technology.explode('skills')

df_technology
```

```python
df_plot = df_da_skills_high_demand.merge(df_technology, left_on= 'job_skills', right_on= 'skills')
```

```python
from adjustText import adjust_text

# fig, ax = plt.subplots()

# df_plot.plot(kind= 'scatter', x= 'skill_percent', y='median_salary')
sns.scatterplot(data= df_plot, x='skill_percent', y='median_salary', hue='technology')

texts = []
for i, txt in enumerate(df_da_skills_high_demand.index):
    texts.append(plt.text(df_da_skills_high_demand['skill_percent'].iloc[i],df_da_skills_high_demand['median_salary'].iloc[i], f"{txt} ({df_da_skills_high_demand['skill_percent'].iloc[i]:.0f}%)"))
    

adjust_text(texts, 
            arrowprops=dict(arrowstyle='-', color='gray'))

plt.xlabel('Percent of Data Analysts Jobs')
plt.ylabel('Median Yearly Salary')
plt.title('Most Optimal Skills for Data Analysts in the Sg')
plt.tight_layout()
plt.xlim(0,60)

for text in texts:
    content = text.get_text()
    if "power bi" in content:  # Check for specific text content
        text.set_position((text.get_position()[0] - 1, text.get_position()[1] - 1500))  # Move text
      
    elif "hadoop" in content:
        text.set_position((text.get_position()[0] - 1, text.get_position()[1] - 1500))

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, post: f'${int(y/1000)}k'))
ax.xaxis.set_major_formatter(PercentFormatter(decimals=0))
```
