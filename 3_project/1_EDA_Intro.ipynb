```python

# Importing Libraries
import ast
import seaborn as sns
import pandas as pd 
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
ds = load_dataset('lukebarousse/data_jobs')
df = ds['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df
```

```python
df_da_sg = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'Singapore')].copy()
df_da_sg 
```

```python
df_plot = df_da_sg['job_location'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data= df_plot, x= 'count', y= 'job_location', hue= 'count', palette='dark:b_r', legend= False)
sns.despine()
plt.title('Number of Data Analyst roles in singapore')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
fig, ax = plt.subplots(1, 3)
fig.set_size_inches(12, 5)

dict_column = {
    'job_work_from_home': 'work from home',
    'job_no_degree_mention': 'Job Degree Req',
    'job_health_insurance': 'Health Insurance Offered'
}

for i, (column, title) in enumerate(dict_column.items()):
    # Get value counts and unique labels dynamically
    value_counts = df_da_sg[column].value_counts()
    labels = value_counts.index.astype(str)  # Convert index to strings
    
    ax[i].pie(value_counts, startangle=90, autopct='%1.1f%%', labels=labels)
    ax[i].set_title(title)

plt.show()   
```

```python
df_plot = df_da_sg['company_name'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data= df_plot, x= 'count', y= 'company_name', hue= 'count', palette='dark:b_r', legend= False)
sns.despine()
plt.title('Number of Data Analyst roles in singapore companies')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```
