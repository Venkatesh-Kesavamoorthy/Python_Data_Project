```python
import pandas as pd 
from datasets import load_dataset
import matplotlib.pyplot as plt

ds = load_dataset('lukebarousse/data_jobs')
df = ds['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.pivot_table(values= 'salary_year_avg', index= 'job_country', columns= 'job_title_short', aggfunc= 'median')
```

```python
tc = df_jc = df['job_country'].value_counts().head(6).index

tc
```

```python
df_jc = df.pivot_table(
    values = 'salary_year_avg',
    index= 'job_country',
    columns= 'job_title_short',
    aggfunc= 'median'
)

df_jc = df_jc.loc[tc]

jt = ['Data Analyst', 'Data Engineer', 'Data Scientist']

df_jc = df_jc[jt]

df_jc.plot(kind= 'bar')
plt.ylabel('median salary in USD')
plt.xlabel('')
plt.title('Median Salary by Job Country and Job Title')
plt.xticks(rotation= 45, ha= 'right')
plt.show()
```
