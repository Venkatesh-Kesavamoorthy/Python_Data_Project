```python
import pandas as pd 
from datasets import load_dataset
import matplotlib.pyplot as plt

ds = load_dataset('lukebarousse/data_jobs')
df = ds['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.iloc[0:10]
```

```python
df.loc[0:10, ['salary_rate', 'salary_year_avg','salary_hour_avg']]
```

```python
median_year_salary = df['salary_year_avg'].median()
```

```python
median_hour_salary = df['salary_hour_avg'].median()
```

```python
df_filled = df

df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_year_salary)

df_filled['salary_year_avg']
```

```python
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_hour_salary)

df_filled['salary_hour_avg']
```

```python
df_unique = df_filled

df_unique = df_filled.drop_duplicates()

print('length of df_unique: ', len(df_unique))
print('length of df_filled: ', len(df_filled))
print('length of duplicates from df_filled - df_unique: ', len(df_filled) - len(df_unique))
```

```python
df_unique = df_filled.drop_duplicates(subset= ['job_title','company_name'])

print('length of df_unique: ', len(df_unique))
print('length of df_filled: ', len(df_filled))
print('length of duplicates from df_filled - df_unique: ', len(df_filled) - len(df_unique))
print('length of df: ', len(df))
```
