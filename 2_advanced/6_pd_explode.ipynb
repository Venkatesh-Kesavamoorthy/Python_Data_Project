```python
import ast
import pandas as pd 
from datasets import load_dataset
import matplotlib.pyplot as plt

ds = load_dataset('lukebarousse/data_jobs')
df = ds['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df
```

```python
df_exploded = df.explode('job_skills')
```

```python
df_exploded
```

```python
df_exploded['job_skills'].value_counts().head(10).plot(kind= 'bar')
```

```python
skl_c = df_exploded.groupby(['job_title_short', 'job_skills']).size()
skl_c
```

```python
df_skill_count = skl_c.reset_index(name= 'skill_count')
df_skill_count = df_skill_count.sort_values(by= 'skill_count', ascending= False)

df_skill_count
```

```python
job_title = 'Data Analyst'
top_skills = 10

df_skill_final = df_skill_count[df_skill_count['job_title_short'] == job_title].head(top_skills)
df_skill_final.plot(kind= 'barh', x = 'job_skills', y = 'skill_count')
plt.gca().invert_yaxis()
plt.title('Top_15_skills_for_Data_Analysts')
plt.legend().set_visible(False)
plt.xlabel('job_count')
plt.ylabel('')
plt.show()
```
